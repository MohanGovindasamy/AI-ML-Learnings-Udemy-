# Polynomial Regression

## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing Dataset

```python
# Read data from data.csv file by using Pandas read_csv() command
data = pd.read_csv('./Position_Salaries.csv')

# print dataset into console
data
```

```python
# Seperate data as Input vs Result
X = data.iloc[ : , 1:-1].values # this will storte all values into X variable expect last column
Y = data.iloc[ : , -1].values # this will storte all last colomn values into y variable

# view dataset into console
X
Y
```

## Training Linear Regression Model on the Whole Sheet

```python
# Import Leanear Regression from sklearn
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()

# Now fit into out training data set
linear_regression.fit(X, Y)
```

## Training Polynomial Regression Model on the Whole Sheet

```python
# Import PolynomialFeatures from sklearn > preprocessing
from sklearn.preprocessing import PolynomialFeatures
poly_regression = PolynomialFeatures(degree=10)

# Fit transform to get x polinomial
x_poly = poly_regression.fit_transform(X)

# now create new linear model for polinomial
new_linear_regression = LinearRegression();

# now fit our X_polinomial into our linear regression with Y data
new_linear_regression.fit(x_poly, Y)
```

## Visualising the Linear Regression Result

```python
# using Scatter plot to visuallising our Linear model
plt.scatter(X, Y, color='red')

# Using line plot to visualising actual vs predected Linear values
plt.plot(X, linear_regression.predict(X), color='green') 

# adding cosmitic to visualising
plt.title("Predected - Level Vs Salary ( Linear Regression )")
plt.xlabel('Levels')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression Result

```python
# using Scatter plot to visuallising our Linear model
plt.scatter(X, Y, color='red')

# Using line plot to visualising actual vs predected Polynomial Regression values
plt.plot(X, new_linear_regression.predict(poly_regression.fit_transform(X)), color='green') 

# adding cosmitic to visualising
plt.title("Predected - Level Vs Salary ( Polynomial Regression )")
plt.xlabel('Levels')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression Result (for Higher resolution and smoother curve)

```python
x_grid = np.arange(min(X), max(X), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(X, Y, color = 'red')
# plt.plot(x_grid, lin_reg_2.predict(x_poly), color= 'blue')
plt.plot(x_grid,  new_linear_regression.predict(poly_regression.fit_transform(x_grid)), color= 'blue')

# adding cosmitic to visualising
plt.title("Predected - Level Vs Salary ( Polynomial Regression )")
plt.xlabel('Levels')
plt.ylabel('Salary')
plt.show()
```
